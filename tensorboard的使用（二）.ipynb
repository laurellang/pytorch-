学习：
1、add_image()函数
2、注意image中数据类型，必须是tensor/numpy，所以如果用PIL会不兼容，要用opencv（numpy类型）
3、从PIL到numpy，需要在add_image()中指定shape中每个数字or维度的含义

```python
from torch.utils.tensorboard import SummaryWriter
import numpy as np
from PIL import Image

root_path = "E:\\小土堆pytorch\\数据集\\hymenoptera_data\\"
label_path = "train\\ants\\5650366_e22b7e1065.jpg"
image_path = root_path + label_path

img = Image.open(image_path)
img_array = np.array(img)
print(type(img_array))

print(img_array.shape)
writer = SummaryWriter("logs2")
writer.add_image("test", img_array, 1, dataformats='HWC')
#y = 2x
for i in range(100):
     writer.add_scalar("y=2x", 2 * i, i)
writer.close()
```
